```python
# 上底・下底・高さを設定（cm）
jou_teihen = 10
ka_teihen = 20
takasa = 5

# 台形の面積を計算して変数に代入
menseki = (jou_teihen + ka_teihen) * takasa / 2

# 数値と文字列を結合して出力
print(str(menseki) + "cm²")
```

```python
# 上底・下底・高さを設定（cm）
top = 10
bottom = 20
height = 5

# 台形の面積の公式を使って計算
# （上辺 + 下辺）× 高さ ÷ 2
area = (top + bottom) * height / 2

# f文字列を使って、計算結果と単位を一緒に表示
print(f"{area}cm²")
```

```python
# 二等辺三角形 底辺×高さ÷2
teihen=20
takasa=20
takasa=2

# 二等辺三角形の面積を変数に代入
menseki=(teihen*takasa/2)

# 数値と文字列を結合して出力
print(str(menseki)+"cm²")
```
